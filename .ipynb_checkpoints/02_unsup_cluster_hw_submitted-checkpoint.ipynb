```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### 1. Reading data
The woldbank_development_2015.csv (can be found in the same folder with this notebook) file contains the World Development Indicators for the 2015 year, downloaded from [The World Bank's webpage](https://databank.worldbank.org/source/world-development-indicators#).


 - Look at the data in any text editor. Build up an overall sense how the data is built up and how is the missing values are represented.
 - Read the file into a pandas dataframe and tell pandas which special pattern means if a value is missing. 
 - The data is in a long format. Convert it into a wide format, where each row is a single country and the columns are the measured features, the Series Codes (eg the first column is 'EG.CFT.ACCS.ZS', the second is 'EG.ELC.ACCS.ZS'. Order of the columns does not matter)!
 - Keep only those rows, which represents counties and NOT regions. Luckily they are well separated in the order they occur!
 - Convert the features to numeric format, which will be needed for modeling!

```python
df_data_raw = pd.read_csv('Data/woldbank_development_2015.csv')
df_data_raw.drop(df_data_raw.tail(1).index,inplace=True) # drop last row

df_data_raw = df_data_raw.replace(to_replace='..',value = np.NaN)
df_data_raw['2015 [YR2015]'] = pd.to_numeric(df_data_raw['2015 [YR2015]'])
df_data = df_data_raw[['Country Name','Series Code','2015 [YR2015]']]

df_data = df_data.pivot_table(index=['Country Name'], columns='Series Code', values='2015 [YR2015]').reset_index()
df_data.columns.name = None
```

```python
df_data.head()
```

```python
df_data_raw.isna().sum()
```

```python
# plt.figure(figsize=(25, 6))
# plt.imshow(df_data.isna())
```

```python
regions = ['World','Europe & Central Asia',
 'Europe & Central Asia (IDA & IBRD countries)',
 'Europe & Central Asia (excluding high income)',
 'European Union','Euro area',
 'Europe & Central Asia',
 'Europe & Central Asia (IDA & IBRD countries)',
 'Europe & Central Asia (excluding high income)',
 'European Union','Late-demographic dividend',
 'Latin America & Caribbean',
 'Latin America & Caribbean (excluding high income)',
 'Latin America & the Caribbean (IDA & IBRD countries)','Low & middle income',
 'Low income',
 'Lower middle income','Middle East & North Africa',
 'Middle East & North Africa (IDA & IBRD countries)',
 'Middle East & North Africa (excluding high income)',
 'Middle income','Post-demographic dividend',
 'Pre-demographic dividend','Russian Federation','Sub-Saharan Africa',
 'Sub-Saharan Africa (IDA & IBRD countries)',
 'Sub-Saharan Africa (excluding high income)']
```

```python
df_data = df_data[~df_data['Country Name'].isin(regions)]
```

# 1. Conclusions:
* the .. character is a sign of missing value, let's replace them with NaN
* I collected only those regions, I could notice.

#### 2. Data preprocessing and inspection

 - Visualize the missing values!
 - Keep only those countries which has less than 700 missing features in the original table and keep features which were missing in less than 20 country in the original table
 - Visualize the missing values again! Now really only a few entries are missing. Impute the missing values with its feature's mean value!
 - How many counties and features do we have left?
 - Read the kept features' descriptions. In the original table the Series Name describe the meaning of the features. What do you think, based only on these information, which counties are the most similar to Hungary? And Norway?  

```python
plt.figure(figsize=(25, 6))
plt.imshow(df_data.isna())
```

```python
df_data.shape
```

```python
df_data['nr_nan_vals'] = df_data.isna().sum(axis=1)
df_data = df_data[df_data['nr_nan_vals'] < 700]
df_data.drop('nr_nan_vals',axis = 'columns',inplace = True)

valid_col_cond = dict(df_data.isna().sum() < 20)
valid_cols = [k for k,v in valid_col_cond.items() if v == True]
print('nr of valid columns',len(valid_cols))
df_data = df_data[valid_cols]
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(df_data.isna())
```

```python
df_data = df_data.fillna(df_data.mean())
```

```python
print('Nr of countries: ',len(df_data))
print('Nr. of features: ', len(df_data.columns))
```

```python
col_codes = dict(zip(df_data_raw[df_data_raw['Series Code'].isin(valid_cols[1:])]['Series Code'],
                     df_data_raw[df_data_raw['Series Code'].isin(valid_cols[1:])]['Series Name']))
col_codes
```

#### 3. PCA
 - Perform PCA with 3 principal components on the filtered, imputed data (from now on, data refers to the filtered, imputed dataset)
 - Plot the three embedded 2D combination next to each other (0 vs 1, 0 vs 2 and 1 vs 2)
 - It seems that the embedding is really dominated by a single direction. Normalize the data (each feature should have zero mean and unit variance after normalization) and re-do the PCA and the plotting (do not delete the previous plots, just make new ones).

```python
from sklearn.decomposition import PCA
```

```python
df_data.head()
```

```python
df_data_3 = df_data.copy()
df_data_3.drop('Country Name',axis='columns',inplace = True)
```

```python
pca = PCA(n_components=3)#, random_state=42)
pca_transformed = pca.fit_transform(df_data_3)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(pca_transformed[:,0],pca_transformed[:,1],'.')
plt.title('0-1')
plt.xlabel('component 0')
plt.ylabel('component 1')

plt.subplot(1,3,2)
plt.plot(pca_transformed[:,0],pca_transformed[:,2],'.')
plt.title('0-2')
plt.xlabel('component 0')
plt.ylabel('component 2')

plt.subplot(1,3,3)
plt.plot(pca_transformed[:,1],pca_transformed[:,2],'.')
plt.title('1-2')
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.tight_layout()
```

```python
df_data_3_norm=(df_data_3-df_data_3.mean())/df_data_3.std()

pca = PCA(n_components=3)#, random_state=42)
pca_transformed_norm = pca.fit_transform(df_data_3_norm)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(pca_transformed_norm[:,0],pca_transformed_norm[:,1],'.')
plt.title('norm 0-1')
plt.xlabel('component 0')
plt.ylabel('component 1')

plt.subplot(1,3,2)
plt.plot(pca_transformed_norm[:,0],pca_transformed_norm[:,2],'.')
plt.title('norm 0-2')
plt.xlabel('component 0')
plt.ylabel('component 2')

plt.subplot(1,3,3)
plt.plot(pca_transformed_norm[:,1],pca_transformed_norm[:,2],'.')
plt.title('norm 1-2')
plt.xlabel('component 1')
plt.ylabel('component 2')

plt.tight_layout()
```

```python
pca.components_
```

```python
pd.DataFrame(pca.components_,columns=df_data_3_norm.columns,index = ['PC-1','PC-2','PC-3'])
```

```python
country_names = list(df_data['Country Name'])
country_name_code_dict = dict(zip(df_data_raw['Country Name'],df_data_raw['Country Code']))
country_code = [country_name_code_dict[n] for n in country_names]
```

```python
plt.figure(figsize=(18,18))
plt.gca().invert_yaxis()
plt.scatter(embedded[:,0],embedded[:,1])
plt.title('Countries')
for i in range(len(embedded)):
    plt.text(embedded[i,0],embedded[i,1],country_names[i])
    if country_names[i] == 'Hungary':
        plt.plot(embedded[i,0],embedded[i,1],'r.',ms=20)
    if country_names[i] == 'Norway':
        plt.plot(embedded[i,0],embedded[i,1],'r.',ms=20)
    
```

```python

Z = hierarchy.linkage(df_data_5_norm, 'ward')
plt.figure(figsize=(23,20))
chategories_5 = hierarchy.dendrogram(Z, leaf_rotation=90, leaf_font_size=8, labels=df_data_5_norm.index)
```
